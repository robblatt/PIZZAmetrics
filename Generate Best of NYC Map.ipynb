```python
import pandas as pd
import folium
from folium import plugins
from folium.features import CustomIcon
```

```python
df = pd.read_csv('final_data.csv', index_col = 0)
```

```python
df.head()
```

# Create "Best Of" Map of NYC

Creates an html file folium map of the overall score

```python
df_ = df.sort_values('overall_score', ascending = False).reset_index(drop = True).copy()
```

```python
# ===================================================
#
# Creates an instance of the map centered on the
# first pizzeria and outputs an HTML file.
# The first five places have a numbered icon
#
# ===================================================

url = 'https://raw.githubusercontent.com/robblatt/PIZZAmetrics/master/images/'
wurl = 'https://raw.githubusercontent.com/robblatt/PIZZAmetrics/master/images/wordclouds/'

icon_image1 = url + 'pizza-1.png'
icon_image2 = url + 'pizza-2.png'
icon_image3 = url + 'pizza-3.png'
icon_image4 = url + 'pizza-4.png'
icon_image5 = url + 'pizza-5.png'
icon_image = url + 'pizza-n.png'

icon1 = CustomIcon(
    icon_image1,
    icon_size=(40, 40),
    icon_anchor=(20, 40),
    popup_anchor=(0, -40)
)

icon2 = CustomIcon(
    icon_image2,
    icon_size=(36, 36),
    icon_anchor=(18, 36),
    popup_anchor=(0, -36)
)

icon3 = CustomIcon(
    icon_image3,
    icon_size=(32, 32),
    icon_anchor=(16, 32),
    popup_anchor=(0, -32)
)

icon4 = CustomIcon(
    icon_image4,
    icon_size=(28, 28),
    icon_anchor=(14, 28),
    popup_anchor=(0, -28)
)

icon5 = CustomIcon(
    icon_image5,
    icon_size=(24, 24),
    icon_anchor=(12, 24),
    popup_anchor=(0, -24)
)

icon = CustomIcon(
    icon_image,
    icon_size=(20, 20),
    icon_anchor=(10, 20),
    popup_anchor=(0, -20)
)

nycMap = folium.Map(location=[df_.laty[0], df_.longy[0]], tiles='stamentoner', zoom_start=12)
folium.TileLayer('openstreetmap').add_to(nycMap)
folium.TileLayer('cartodbdark_matter').add_to(nycMap)
folium.LayerControl().add_to(nycMap)

for i in range(0, 5):
    html = '''<h3>''' + df_.name[i] + '''</h3><br/> <img src="''' + wurl + df_.name[i][:6] + '''.png">'''
    if i == 0:
        marker = folium.Marker(
            location=[df_.laty[i],df_.longy[i]],
            icon=icon1,
            popup=html,
        ).add_to(nycMap)
    elif i == 1:
        marker = folium.Marker(
            location=[df_.laty[i],df_.longy[i]],
            icon=icon2,
            popup=html,
        ).add_to(nycMap)
    elif i == 2:
        marker = folium.Marker(
            location=[df_.laty[i],df_.longy[i]],
            icon=icon3,
            popup=html,
        ).add_to(nycMap)
    elif i == 3:
        marker = folium.Marker(
            location=[df_.laty[i],df_.longy[i]],
            icon=icon4,
            popup=html,
        ).add_to(nycMap)
    elif i == 4:
        marker = folium.Marker(
            location=[df_.laty[i],df_.longy[i]],
            icon=icon5,
            popup=html,
        ).add_to(nycMap)

for i in range(5,26):
    html = '''<h3>''' + df_.name[i] + '''</h3><br/> <img src="''' + wurl + df_.name[i][:6] + '''.png">'''
    folium.Marker(location = [df_.iloc[i]['laty'], df_.iloc[i]['longy']], 
                  popup=html, 
                  icon = CustomIcon(
                            icon_image,
                            icon_size=(20, 20),
                            icon_anchor=(10, 20),
                            popup_anchor=(0, -20)
                        )
                 ).add_to(nycMap)

for i in range(26,len(df_)):
    html = '''<h3>''' + df_.name[i] + '''</h3><br/> <img src=''' + wurl + df_.name[i][:6] + '''.png">'''
    folium.CircleMarker(
        (df_.laty[i],df_.longy[i]), 
        radius=5, weight=1, color='red', 
        fill_color='red', fill_opacity=.3,
        popup = html
    ).add_to(nycMap)

nycMap.save('overall.html')    
nycMap
```
