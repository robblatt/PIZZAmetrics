```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt

import datetime
```

```python
csv_file1 = 'csv/l-and-b-spumoni-gardens-brooklyn reviews.csv'
csv_file2= 'csv/lombardis-pizza-new-york-4 reviews.csv'
```

```python
def reviews_over_time(csv_file, the_title):

    df = pd.read_csv(csv_file, index_col = 0)

    # clean dates

    user_dates = df.user_date.copy()

    sorted_user_dates = [None] * len(df.user_date)
    for i in range(len(user_dates)):
        sorted_user_dates[i] = re.sub(r"[a-zA-z]+", "", str(df.user_date[i]))

    df.user_date = pd.to_datetime(sorted_user_dates)
    df.sort_values('user_date', inplace = True)
    df.reset_index(inplace = True, drop = True)

    df.head()

    mean_rating = []

    for i in range(len(df.stars)):
        if i > 0:
            mean_rating.append(sum(df.stars[0:i])/(i))
        else:
            mean_rating.append(df.stars[i])
            
            
    df['mean_rating'] = mean_rating
    
    df['roll_mean'] = df.stars.rolling(window=500, min_periods=1).mean()
    
    k_reviews = [None] * len(df)
    for i in range(len(k_reviews)):
        k_reviews[i] = i // 1000

    df['k_reviews'] = k_reviews

    df.reset_index(inplace = True)

    date_df = df.groupby(['user_date']).max()

    sam_mean = sum(df.stars)/len(df.stars)

    r1 = 0
    r2 = len(date_df.stars)
    ep = 0.1

    fig = plt.figure(figsize=(12, 6))
    ax = fig.subplots()

    for i in range(len(date_df)):
        if round(sam_mean*2)/2 - .25 < date_df.mean_rating[i] < round(sam_mean*2)/2 + .25:
            if i < convergence_point:
                convergence_point = i
        else:
            convergence_point = 900000

    converge_date = date_df.iloc[convergence_point].name
    
    plt.rcParams['axes.facecolor']='#89BBB6'
            
    plt.axvline(x=converge_date, color = 'w', linewidth=1.0)
    
    plt.text(converge_date, .95, 'Convergence Date: ' + str(converge_date.date()), transform=ax.get_xaxis_transform())

    for i in range(date_df['k_reviews'].min(), date_df['k_reviews'].max() + 1):
        plt.axvline(x=date_df[date_df['k_reviews'] == i].iloc[0].name, color = '#EDA939', linewidth=1.0)
        
    ax.axhline(y=4.75, color='k', linestyle=(0, (9, 10)), linewidth=1.0)
    ax.axhline(y=4.25, color='k', linestyle=(0, (5, 10)), linewidth=1.0)
    ax.axhline(y=3.75, color='k', linestyle=(0, (1, 10)), linewidth=1.0)
    plt.text(0.3, 0.86, '5 star threshold', transform=ax.transAxes)
    plt.text(0.3, 0.58, '4.5 star threshold', transform=ax.transAxes)
    plt.text(0.3, 0.3, '4 star threshold', transform=ax.transAxes)
    plt.text(0.3, 0.01, 'Yellow lines denote every 1,000 reviews', transform=ax.transAxes)

    ax.axhline(y=sam_mean, color='g', linestyle='-', label = 'mean review score: ' + str(round(sam_mean, 4)), linewidth=3.0)
    ax.set(xlabel='Date of Reviews', ylabel='Mean Star Ratings on Yelp', title= the_title)
    ax.plot(date_df.roll_mean, label = 'Rolling Mean (250 Reviews)', linewidth=3.0, color='#C74223')
    ax.plot(date_df.mean_rating, label = 'Review Mean', color = '#343434')

    ax.legend(frameon = False)

    ax.set_xlim(date_df.index.min(), date_df.index.max());
    ax.set_ylim(3.25, 5.01)
```

```python
def compare_annual_reviews(csv1, csv2):
    """Was built to compare L&B Spumoni Gardens and 
    Lombardi's annual Yelp reviews. There is a one year
    discrepancy between then (2005), so a blank year is
    added to L&B.
    """

    df1 = pd.read_csv(csv1, index_col = 0)
    df2 = pd.read_csv(csv2, index_col = 0)

    user_dates1 = df1.user_date.copy()
    user_dates2 = df2.user_date.copy()

    sorted_user_dates1 = [None] * len(df1.user_date)
    for i in range(len(user_dates1)):
        sorted_user_dates1[i] = re.sub(r"[a-zA-z]+", "", str(df1.user_date[i]))

    sorted_user_dates2 = [None] * len(df2.user_date)
    for i in range(len(user_dates2)):
        sorted_user_dates2[i] = re.sub(r"[a-zA-z]+", "", str(df2.user_date[i]))

    df1.user_date = pd.to_datetime(sorted_user_dates1)
    df1.sort_values('user_date', inplace = True)
    df1.reset_index(inplace = True, drop = True)

    df2.user_date = pd.to_datetime(sorted_user_dates2)
    df2.sort_values('user_date', inplace = True)
    df2.reset_index(inplace = True, drop = True)

    mean_rating1 = []
    mean_rating2 = []

    for i in range(len(df1.stars)):
        if i > 0:
            mean_rating1.append(sum(df1.stars[0:i])/(i))
        else:
            mean_rating1.append(df1.stars[i])

    for i in range(len(df2.stars)):
        if i > 0:
            mean_rating2.append(sum(df2.stars[0:i])/(i))
        else:
            mean_rating2.append(df2.stars[i])


    df1['mean_rating'] = mean_rating1
    df2['mean_rating'] = mean_rating2

    df1['roll_mean'] = df1.stars.rolling(window=1000, min_periods=1).mean()
    df2['roll_mean'] = df2.stars.rolling(window=1000, min_periods=1).mean()

    counting1 = df1.user_date.groupby([df1.user_date.dt.to_period("Y")]).agg({'count'})
    counting1['mean_rating'] = df1.mean_rating.groupby([df1.user_date.dt.to_period("Y")]).max().to_frame()['mean_rating']

    counting2 = df2.user_date.groupby([df2.user_date.dt.to_period("Y")]).agg({'count'})
    counting2['mean_rating'] = df2.mean_rating.groupby([df2.user_date.dt.to_period("Y")]).max().to_frame()['mean_rating']

    counting1.reset_index(inplace = True)
    counting2.reset_index(inplace = True)

    counting1 = counting1.append({'user_date': pd.Period(freq ='T', year = 2005, month = 10, hour = 15, minute = 49).asfreq('A-DEC'), 'count' : 0, 'mean_rating' : 5}, ignore_index=True)

    counting1.reset_index(inplace = True, drop = True)
    counting1.sort_values('user_date', inplace = True)
    counting1.reset_index(inplace = True, drop = True)

    counting2.reset_index(inplace = True, drop = True)
    counting2.sort_values('user_date', inplace = True)
    counting2.reset_index(inplace = True, drop = True)

    plt.rcParams['axes.facecolor']='#89BBB6'

    bars = plt.figure(1)
    bars.set_figheight(6)
    bars.set_figwidth(12)

    width = 0.4
    label = counting1.user_date

    ax = bars.add_subplot(1,1,1)

    ax.bar(counting2.index - width/2, counting2['count'], width, color = 'C74223', label = "Lombardi's")
    ax.bar(counting1.index + width/2, counting1['count'], width, color = '#343434', label = "L&B Spumoni Garden")
    plt.xticks(counting1.index, label, fontsize=10, rotation=30)
    plt.title('Yelp Reviews per Year')
    plt.legend(frameon = False)
```

```python
compare_annual_reviews(csv_file1, csv_file2)
```
