# Exploring How Yelp Reviews Change

```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
def large_numbers(csv_file):
    
    df = pd.read_csv(csv_file, index_col = 0)

    # create a list of the dates, remove some text that found its way into the dates

    user_dates = df.user_date.copy()

    sorted_user_dates = [None] * len(df.user_date)
    for i in range(len(user_dates)):
        sorted_user_dates[i] = re.sub(r"[a-zA-z]+", "", str(df.user_date[i]))

    df.user_date = pd.to_datetime(sorted_user_dates)
    df.sort_values('user_date', inplace = True)
    df.reset_index(inplace = True, drop = True)

    # creates a list of the review mean after each new review

    mean_rating = []

    for i in range(len(df.stars)):
        if i > 0:
            mean_rating.append(sum(df.stars[0:i])/(i))
        else:
            mean_rating.append(df.stars[i])

    r1 = 0
    r2 = len(df.stars)
    ep = 0.1
    sam_mean = sum(df.stars)/len(df.stars)

    divi = len(df) // 6
    
    xticks = [(df.user_date[i * divi].date()) for i in range(0,7)]

    fig = plt.figure(figsize=(15, 8))
    ax = fig.subplots()

    ax.set(xlabel='Date of Reviews', ylabel='Mean Star Ratings on Yelp',
           title='The mean of ' + csv_file + "'s reviews")
    ax.axhline(y=sam_mean, color='r', linestyle='-', label = 'mean review score', linewidth=3.0)
    ax.plot(mean_rating, label = 'Review Mean')
    ax.plot(df.stars.rolling(window=1000, min_periods=10).mean(), label = 'Rolling Mean (500)')
    ax.grid(b = True, axis = 'y')
    ax.legend()

    ax.set_xlim(r1, r2)
    ax.set_ylim(df.stars.rolling(window=1000, min_periods=10).mean().min(),df.stars.rolling(window=1000, min_periods=10).mean().max())
    ax.set_xticklabels(xticks)
```

```python
large_numbers('csv/lombardis-pizza-new-york-4 reviews.csv')
```

```python
large_numbers('csv/l-and-b-spumoni-gardens-brooklyn reviews.csv')
```
